```python
df2 = df[(df['Event distance/length'].isin(['50mi','50km'])) & (df['Year of event'] == 2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
df2.head(10)
```

```python
df2.shape
```

```python
#remove (USA) from event name
```

```python
df2['Event name'].str.split('(').str.get(0)
```

```python
df2['Event name'] = df2['Event name'].str.split('(').str.get(0)
```

```python
df2.head(10)
```

```python
#clean up athlete age
```

```python
df2['athlete age'] = 2020 - df2['Athlete year of birth']
```

```python
#remove h from athlete performance
```

```python
df2['Athlete performance'].str.split('h').str.get(0)
```

```python
df2['Athlete performance'] = df2['Athlete performance'].str.split('h').str.get(0)
```

```python
df2.head(10)
```

```python
#drop columns :athlete club, athlete country, athlete yob, athlete age category
```

```python
df3 = df2.drop(['Athlete club', 'Athlete country', 'Athlete year of birth', 'Athlete age category'], axis = 1)
```

```python
df3.head(10)
```

```python
df3.shape
```

```python
#clean up null values
```

```python
df3.isna().sum()
```

```python
df3[df3['athlete age'].isna()==1]
```

```python
df4 = df3.dropna()
```

```python
df4.shape
```

```python
#check for dupes
```

```python
df4[df2.duplicated() == True]
```

```python
#reset index
```

```python
df4.reset_index(drop = True)
```

```python
#fix types
```

```python
df4.dtypes
```

```python
df4['athlete age'] = df4['athlete age'].astype(int)
```

```python
df4['Athlete average speed'] = df4['Athlete average speed'].astype(float)
```

```python
df4.dtypes
```

```python
df4.head(10)
```

```python
#rename columns
```

```python
#Year of event                  int64
#Event dates                   object
#Event name                    object
#Event distance/length         object
#Event number of finishers      int64
#Athlete performance           object
#Athlete gender                object
#Athlete average speed        float64
#Athlete ID                     int64
#athlete age                    int32
```

```python
df4 = df4.rename(columns = {'Year of event': 'year' ,
                            'Event dates': 'race_day' ,
                            'Event name': 'race_name',
                            'Event distance/length': 'race_length',
                            'Event number of finishers': 'race_number_of_finishers',
                            'Athlete performance': 'athlete_performance',
                            'Athlete gender': 'athlete_gender',
                            'Athlete average speed': 'athlete_average_speed',
                            'Athlete ID': 'athlete_id',
                            'athlete age': 'athlete_age'})
```

```python
df4.head(10)
```

```python
#reorder columns
```

```python
df4 = df4[['race_day', 'race_name', 'race_length', 'race_number_of_finishers', 'athlete_id', 'athlete_gender','athlete_age','athlete_performance','athlete_average_speed']]
```

```python
df4['race_month'] = df4['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df4['race_season'] = df4['race_month'].apply(lambda x: 'Winter' if x > 11 else 'Fall' if x>8 else 'Summer' if x > 5 else 'spring' if x > 2 else 'Winter') 
```
